### Loading the webpage
First we import the packages

````python
from bs4 import BeautifulSoup
import requests
import csv
print('Modules imported :)')
````

The webpage we are gooing to srape is the [wiki page](https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M) List of postal codes of Canada


````python
uri = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
'The URI is: {} '.format(uri) 
````

Now we make the soup

````python
source = requests.get(uri).text
soup = BeautifulSoup(source, 'lxml')
````

````python
table = soup.find('table') 
print(table.prettify())
````
